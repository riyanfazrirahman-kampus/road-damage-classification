```python
import os

FOLDER_PATH = "./dataset/local/tidak_rusak"  # ganti ke folder yang ingin diubah

for root, dirs, files in os.walk(FOLDER_PATH):
    for filename in files:
        # hanya ubah jika nama file diawali "Screenshot"
        if filename.startswith("Screenshot"):
            old_path = os.path.join(root, filename)

            new_filename = filename.replace("Screenshot", "IMG", 1)
            new_path = os.path.join(root, new_filename)

            os.rename(old_path, new_path)
            print(f"{filename} -> {new_filename}")

print("Rename selesai ✅")
```

```python

import os

FOLDER_PATH = "./dataset/public/alur"  # ganti sesuai folder

for root, dirs, files in os.walk(FOLDER_PATH):
    for filename in files:
        if "aug" in filename:
            old_path = os.path.join(root, filename)

            # hapus semua tulisan "aug"
            new_filename = filename.replace("aug", "")

            # rapikan jika jadi ada "__" atau "_."
            new_filename = new_filename.replace("__", "_")
            new_filename = new_filename.replace("_.", ".")

            new_path = os.path.join(root, new_filename)

            os.rename(old_path, new_path)
            print(f"{filename} -> {new_filename}")

print("Semua kata 'aug' berhasil dihapus dari nama file ✅")
```

```python
import os

folder_path = "../dataset/local/alur"

total = 0
asli = 0
aug = 0

for file in os.listdir(folder_path):
    if os.path.isfile(os.path.join(folder_path, file)):
        total += 1
        if "_aug_" in file:
            aug += 1
        else:
            asli += 1

print("Total file :", total)
print("Data asli  :", asli)
print("Data aug   :", aug)
```

```python
import os

folder_path = "../dataset/public/tidak_rusak"

deleted = 0

for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    
    if os.path.isfile(file_path) and "_aug" in file:
        os.remove(file_path)
        deleted += 1

print(f"File aug yang dihapus: {deleted}")
```

```python
import os
import random
from PIL import Image, ImageEnhance

def augment_6x_per_image(dataset_dir):
    AUG_TYPES = ["flip", "flip_v", "bright", "dark", "zoom", "rotate"]

    images = [
        f for f in os.listdir(dataset_dir)
        if f.lower().endswith(('.jpg', '.png', '.jpeg'))
    ]

    if not images:
        raise ValueError("Tidak ada gambar di folder!")

    print(f"Total gambar asli: {len(images)}")

    for img_name in images:
        img_path = os.path.join(dataset_dir, img_name)
        base_name, ext = os.path.splitext(img_name)

        img = Image.open(img_path).convert("RGB")

        for aug_type in AUG_TYPES:
            aug_img = img.copy()

            if aug_type == "flip":
                aug_img = aug_img.transpose(Image.FLIP_LEFT_RIGHT)

            elif aug_type == "flip_v":
                aug_img = aug_img.transpose(Image.FLIP_TOP_BOTTOM)

            elif aug_type == "bright":
                aug_img = ImageEnhance.Brightness(aug_img).enhance(1.7)

            elif aug_type == "dark":
                aug_img = ImageEnhance.Brightness(aug_img).enhance(0.3)

            elif aug_type == "zoom":
                w, h = aug_img.size
                z = random.uniform(1.3, 2)
                nw, nh = int(w/z), int(h/z)
                left, top = (w-nw)//2, (h-nh)//2
                aug_img = aug_img.crop((left, top, left+nw, top+nh)).resize((w, h))

            elif aug_type == "rotate":
                aug_img = aug_img.rotate(random.randint(-20, 20), fillcolor=(128,128,128))

            new_name = f"{base_name}_aug_{aug_type}{ext.lower()}"
            aug_img.save(os.path.join(dataset_dir, new_name))

    print("Augmentasi 6x per gambar selesai ✅")

augment_6x_per_image("../dataset/local/retak") 
```
