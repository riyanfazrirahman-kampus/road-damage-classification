```python
import os
import shutil
```

```python

# ===== PATH =====
BASE_MODEL = "../../model-classification"
BASE_API = "../../model-fastapi/app"

FROM_DIR = os.path.join(BASE_MODEL, "models")
TO_DIR = os.path.join(BASE_API, "models")

# Pastikan folder tujuan ada
os.makedirs(TO_DIR, exist_ok=True)
```

```python

# ===== PROSES COPY =====
for model_name in os.listdir(FROM_DIR):
    src_model_path = os.path.join(FROM_DIR, model_name)

    # Skip kalau bukan folder
    if not os.path.isdir(src_model_path):
        continue

    dst_model_path = os.path.join(TO_DIR, model_name)
    os.makedirs(dst_model_path, exist_ok=True)

    # File yang mau diambil
    files_to_copy = ["labels.json", "model.h5"]

    for file_name in files_to_copy:
        src_file = os.path.join(src_model_path, file_name)
        dst_file = os.path.join(dst_model_path, file_name)

        if os.path.exists(src_file):
            shutil.copy2(src_file, dst_file)
            print(f"[✓] Copied: {src_file} -> {dst_file}")
        else:
            print(f"[!] File tidak ditemukan: {src_file}")

print("\nSelesai copy semua model!")
```
