```python
# ==========================================
# SPLIT DATASET (80:10:10)
# INPUT  : dataset_aug1
# OUTPUT : dataset_split/
#           ├── train/
#           ├── val/
#           └── test/
# ==========================================

import os
import shutil
import random
from tqdm import tqdm
```

```python
# Path
INPUT_DIR = "../dataset_merged_aug"
OUTPUT_DIR = "../dataset_split_aug"

SPLIT_RATIO = {
    "train": 0.8,
    "val": 0.1,
    "test": 0.1
}

random.seed(42)
```

```python

# ==========================================
# AMBIL KELAS OTOMATIS
# ==========================================

def get_classes(dir_path):
    return sorted([
        d for d in os.listdir(dir_path)
        if os.path.isdir(os.path.join(dir_path, d))
    ])

CLASSES = get_classes(INPUT_DIR)
print("Kelas:", CLASSES)

# BUAT FOLDER OUTPUT
for split in SPLIT_RATIO.keys():
    for cls in CLASSES:
        os.makedirs(os.path.join(OUTPUT_DIR, split, cls), exist_ok=True)

# SPLIT PROSES
for cls in CLASSES:
    cls_path = os.path.join(INPUT_DIR, cls)
    images = [f for f in os.listdir(cls_path) if f.lower().endswith((".jpg", ".png", ".jpeg"))]

    random.shuffle(images)

    total = len(images)
    n_train = int(total * SPLIT_RATIO["train"])
    n_val = int(total * SPLIT_RATIO["val"])

    train_files = images[:n_train]
    val_files = images[n_train:n_train + n_val]
    test_files = images[n_train + n_val:]

    print(f"\n{cls} -> total: {total} | train: {len(train_files)}, val: {len(val_files)}, test: {len(test_files)}")

    # Copy file
    for split_name, file_list in zip(["train", "val", "test"], [train_files, val_files, test_files]):
        for f in tqdm(file_list, desc=f"Copy {cls} -> {split_name}"):
            src = os.path.join(cls_path, f)
            dst = os.path.join(OUTPUT_DIR, split_name, cls, f)
            shutil.copy2(src, dst)

print("\nSelesai split dataset!")

# ==========================================
# CEK JUMLAH AKHIR
# ==========================================

print("\nJumlah akhir per split:")
for split in ["train", "val", "test"]:
    print(f"\n{split.upper()}")
    for cls in CLASSES:
        path = os.path.join(OUTPUT_DIR, split, cls)
        total = len(os.listdir(path))
        print(f"{cls}: {total}")
```
