```python
# ==========================================
# MERGE DATASET LOCAL & PUBLIK (AUTO CLASS)
# Ambil nama kelas langsung dari folder
# ==========================================

import os
import shutil
from tqdm import tqdm
```

```python

# Path dataset
BASE_DIR = "../dataset"
LOCAL_DIR = os.path.join(BASE_DIR, "local")
PUBLIK_DIR = os.path.join(BASE_DIR, "public")
OUTPUT_DIR = os.path.join("../", "dataset_merged")

# Ambil kelas otomatis dari folder local (atau publik)
def get_classes(*dirs):
    classes = set()
    for d in dirs:
        if os.path.exists(d):
            classes.update([
                name for name in os.listdir(d)
                if os.path.isdir(os.path.join(d, name))
            ])
    return sorted(list(classes))

CLASSES = get_classes(LOCAL_DIR, PUBLIK_DIR)
print("Kelas ditemukan:", CLASSES)

# Buat folder output
for cls in CLASSES:
    os.makedirs(os.path.join(OUTPUT_DIR, cls), exist_ok=True)

# Fungsi copy dengan rename biar gak bentrok

def copy_images(src_dir, dst_dir, prefix):
    if not os.path.exists(src_dir):
        return

    files = [f for f in os.listdir(src_dir) if f.lower().endswith((".jpg", ".png", ".jpeg"))]
    
    for i, file in enumerate(tqdm(files, desc=f"Copy {prefix}")):
        src_path = os.path.join(src_dir, file)
        
        # Rename biar unik
        ext = os.path.splitext(file)[1]
        new_name = f"{prefix}_{i}{ext}"
        dst_path = os.path.join(dst_dir, new_name)
        
        shutil.copy2(src_path, dst_path)

# Proses merge
for cls in CLASSES:
    print(f"\nProcessing class: {cls}")
    
    local_path = os.path.join(LOCAL_DIR, cls)
    publik_path = os.path.join(PUBLIK_DIR, cls)
    output_path = os.path.join(OUTPUT_DIR, cls)
    
    copy_images(local_path, output_path, prefix="local")
    copy_images(publik_path, output_path, prefix="public")

print("\nSelesai! Dataset sudah digabung di folder 'merged'")

# ==========================================
# (Optional) CEK JUMLAH DATA
# ==========================================

print("\nJumlah data per kelas:")
for cls in CLASSES:
    path = os.path.join(OUTPUT_DIR, cls)
    total = len(os.listdir(path))
    print(f"{cls}: {total} gambar")
```
